```python

def to_mne(cml_eeg): 
    """
    Convert data to MNE's RawArray or EpochsArray format. If loading
    continuous data, a RawArray will be returned. If loading epoched data,
    an EpochsArray will be returned. Events are accessible made accessible
    through the 'events' field of the RawArray/EpochsArray's info.

    If the data was originally loaded into the EEGContainer using MNE, the
    returned MNE object will retain the original info structure created by
    mne.io.read_raw_*. Otherwise, a default info object will be attached to
    the RawArray/EpochsArray.

    :return: A RawArray or EpochsArray constructed from the EEGContainer's
            data.
    """
    import mne
    
    cml_eeg.data = cml_eeg.data.astype('float32')
    
    
    
    # Create info
    info = mne.create_info(
        ch_names = [str(c) for c in cml_eeg.channels],
        sfreq = cml_eeg.samplerate, 
        ch_types = 'eeg'
    )
    

    # Return RawArray if loading continuous data (must be from from a
    # single recording)
    if len(cml_eeg.epochs) == 1 and \
            np.all(cml_eeg.epochs[0] == np.array([0, None])):
        eeg = mne.io.RawArray(cml_eeg.data[0], info, first_samp=0)
    # Return EpochsArray if loading epoched data
    else:
        chunked_events = np.array_split(cml_eeg.data, 100, axis = 0)
        chunk_eegs = [] 
        for chunk in chunked_events:
            ceeg = mne.EpochsArray(chunk, info, tmin = cml_eeg.time[0] / 1000 )
            chunk_eegs.append(ceeg)
            
        eeg = mne.concatenate_epochs(chunk_eegs)
        
        #eeg = mne.EpochsArray(cml_eeg.data, info, tmin=cml_eeg.time[0] / 1000.)

    # Attach events to MNE object as record array
    if cml_eeg.events is not None:
        events = cml_eeg.events.to_records()
    else:
        columns = ["eegoffset", "epoch_end"]
        if len(cml_eeg.epochs[0]) > 2:
            columns = [columns[i] if i < 2 else "column_{}".format(i)
                       for i in range(len(cml_eeg.epochs[0]))]
        events = pd.DataFrame(cml_eeg.epochs, columns=columns).to_records(
            index=False)
    #eeg.info['events'] = events

    return eeg
```
