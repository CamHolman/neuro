```python
from ptsa.data.filters import MorletWaveletFilter
import numpy as np

def compute_phase(eeg, freq, buffer_len):
        # compute phase
        MWF = MorletWaveletFilter(np.array([freq]),
                                        output='phase',
                                        width=5,
                                        cpus=12,
                                        verbose=False)

        # remove the buffer from each end
        phase = MWF.filter(timeseries = eeg)
        #phase_data = phase_data.remove_buffer(buffer_len)
        return phase[0]
```
